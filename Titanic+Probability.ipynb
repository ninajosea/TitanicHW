```python
#hasfamily: 0 or 1
import numpy as np

df['hasfamily'] = np.where(df['sibsp'] + df['parch'] > 0, "1", "0")
df.head(100)
```

```python
#Gender
import numpy as np

df['gender'] = np.where(df['sex'] == 'female', "1", "0")
df.head(100)
```

```python
#Accommodation
import numpy as np

df['accommodation'] = np.where(df['cabin'].isnull(), "0", "1")

df.head(100)
```

```python
femaleSurvived = len((df['gender'] == 1) & (df['survived'] == 1))
familySurvived = len((df['hasfamily'] == 1) & (df['survived'] == 1))
accommodationSurvived = len((df['accommodation'] == 1) & (df['survived'] == 1))
survived = len(df['survived'] == 1)

p_female_survived = femaleSurvived / survived
p_family_survived = familySurvived / survived
p_accommodation_survived = accommodationSurvived / survived

```

```python
def prob_s_fhc(x, y, z):
    return x * y * z
```

```python
prob_s_fhc(p_female_survived, p_family_survived, p_accommodation_survived)
```
